# Let's talk about regularization

```python
df.head()
```

```python
x = df.drop(['Profit'], axis=1)
y = df['Profit']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state = 0)
```

```python
xtrain.head()
```

# L1 Regularization with Lasso

```python
lasso_model = Lasso()
lasso_model.fit(xtrain, ytrain)
```

```python
ridge_model = Ridge()
ridge_model.fit(xtrain, ytrain)
```

```python
lasso_model.score(xtrain, ytrain)
```

```python
lasso_model.score(xtest, ytest)
```

```python
ridge_model.score(xtrain, ytrain)
```

```python
ridge_model.score(xtest, ytest)
```

```python
lasso_prediction = lasso_model.predict(xtest)
lasso_mse = mean_squared_error(ytest, lasso_prediction)
lasso_mse
```

```python
ridge_predictor = ridge_model.predict(xtest)
ridge_mse = mean_squared_error(ytest, ridge_predictor)
ridge_mse
```
